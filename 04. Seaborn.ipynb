# A) Distribution Plots

Some plots that allow us to visualize the distribution of a data set:

* distplot
* jointplot
* pairplot
* rugplot
* kdeplot

## Imports

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

## 1. distplot

The distplot shows the distribution of a univariate set of observations.

```python
sns.distplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.histplot(tips['total_bill'])
```

To remove the kde layer and just have the histogram use:

```python
sns.distplot(tips['total_bill'],kde=False)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=30)
#bins -> divides x axis -> defines 30 points
```

## 2. jointplot

jointplot() allows us to basically match up two distplots for bivariate data. With our choice of what **kind** parameter to compare with: 
* “scatter” 
* “reg” 
* “resid” 
* “kde” 
* “hex”

<b>JOINT SCATTER PLOT

Default for ‘kind’ is scatter.<br>So we do not need to write the kind of the graph if we want to draw a scatter plot.<br>
We see the distributions of total bill and tip on the margins and we see the correlation of them on the ‘ joint’ at the center.

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

Trend - As we go higher in total_bill, we go higher in 'tip'.

<b>REGRESSION PLOT
    
Linear regression attempts to model the relationship between two variables by fitting a linear equation to observed data. One variable is considered to be an explanatory variable, and the other a dependent variable. 
    
If we want to model the correlation of the tips given to the servers and the total bill of the customers we use regression plots to interpret the data much better.<br>The shade around the line shows the confidence level.<br>This level can be changed by changing keyword arguments of regplot() which is used in the jointplot graph to draw the regression lines.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

<b>KERNEL DENSITY ESTIMATION PLOT
    
Kernel density estimation(KDE) is a useful technique to create a smooth curve given a set of data.<br>But KDE is not only the shape of data, but also a continous displacement for the discrete histogram.<br>It can also be used to generate points that look like they came from a certain dataset - this behavior can power simple simulations, where simulated objects are modeled off of real data.

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='kde') 
```

<b>RESIDUAL PLOT
    
A residual is a measure of how well a line fits an individual data point.<br> For data points above the line, the residual is positive, and for data points below the line, the residual is negative.<br> Checking residuals is important to make the proper regression model.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
```

<b>HEXAGON PLOT
    
Hexagonal binned scatter plot(hexbin) is a fairly uncommon style of graphing. Hexbin plots take in lists of X and Y values and return what looks somewhat similar to a scatter plot. <br>In a hexbin the entire graphing space has been divided into hexagons and all points have been grouped into their respective hexagonal regions and color tones are added to indicate the density of each gradient indicating the density of each hexagonal area. <br>White hexagons mean there is no data at that point while dark hexagons mean there is a high density at those points. <br>Hexplots are not very common but they are easier to read than scatter plots.

```python
sns.jointplot(x='total_bill',y='tip',data=tips, kind='hex')
```

# B) Categorical Plots

These help us see the distribution of a categorical column with another categorical or numerical column.

## 1. Barplot

A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean.<br>It can also be understood as a visualization of the group by action. <br>To use this plot we choose a categorical column for the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column.

<b>Syntax: barplot([hue, x, y, data, order, hue_order, ...])

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

<b>By default</b>, it shows us the <b>mean</b> of the total bill per categorical column value.<br>Looking at the plot, we can say that the <b>average total_bill</b> for the male is more as compared to the female. 

<b>● palette</b> is used to set the color of the plot<br>
<b>● estimator</b> is used as a statistical function for estimation within each categorical bin.

```python
sns.barplot(x='sex', y='total_bill', data=tips, estimator=np.std)
```

This tells us the <b>standard deviation per category.</b>

## 2. Countplot

A countplot basically counts the categories and returns a count of their occurrences. <br>It is one of the most simple plots provided by the seaborn library.

<b>Syntax: countplot([x, y, hue, data, order, …])

```python
sns.countplot(x='sex', data=tips)
```

As it only returns the count based off a categorical column, we need to specify only the x parameter.

## 3. Boxplot / the box / Whisker plot

It shows the distribution of the quantitative data that represents the comparisons between variables.<br>boxplot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution i.e. the dots indicating the presence of outliers.

<b>Syntax: boxplot([x, y, hue, data, order, hue_order, …])

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

x takes the categorical column and y is a numerical column.<br>Hence we can see the total bill spent each day.

```python
sns.boxplot(x='day', y='total_bill', data=tips, hue='smoker')
```

<b>"hue"</b> parameter is used to further add a categorical separation.<br>
By looking at the plot we can say that the people who do not smoke had a higher bill on Friday as compared to the people who smoked.

## 4. Violinplot

It is similar to the boxplot except that it provides a higher, more advanced visualization and <b>uses the kernel density estimation</b> to give a better description about the data distribution. 

<b>Syntax: violoinplot([x, y, hue, data, order, ...])

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

Now, the violin plot unlike the boxplot allows us to actually plot all the components that correspond to actual data points.<br>
And it's essentially showing the kernel distribution estimation of the underlying distribution.

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

<b>● hue</b> is used to separate the data further using the sex category<br>
<b>● setting split = True</b> will draw half of a violin for each level. This can make it easier to directly compare the distributions.

## 5. Stripplot

It basically creates a scatter plot based on the category.

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

<b>● One problem</b> with strip plot is that you can’t really tell which points are stacked on top of each other and hence we use the <b>jitter parameter</b> to add some random noise.<br>
<b>● jitter parameter</b> is used to add an amount of jitter (only along the categorical axis) which can be useful when you have many points and they overlap, so that it is easier to see the distribution.<br>
<b>● hue</b> is used to provide an addition categorical separation<br>
<b>● setting dodge = True</b> is used to draw separate strip plots based on the category specified by the hue parameter.

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex')
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', dodge=True)
```

## 6. Swarmplot

It is very similar to the stripplot except the fact that the <b>points are adjusted so that they do not overlap.<br>

```python
sns.swarmplot(x='day', y='total_bill', data=tips)
```

<b>One drawback</b> to using swarmplot is that sometimes they <b>don't scale well</b> to really large numbers and takes a lot of computation to arrange them. <br>So in case we want to visualize a swarmplot properly, we can plot it on <b>top of a violinplot.</b>

```python
sns.violinplot(x='day', y='total_bill', data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='black')
```

## 7. Factorplot -> Catplot (now)

It is the most general of all these plots and provides a parameter called <b>kind</b> to choose the kind of plot we want, thus saving us from the trouble of writing these plots separately. The kind parameter can be bar, violin, swarm etc.

```python
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

# C) Matrix Plots

A matrix plot is an array of scatterplots!

## 1. Heatmap

Heatmap is a way to show some sort of matrix plot.<br>To use a heatmap, the data should be in a matrix form.<br>By matrix, we mean that the index name and the column name must match in some way, so that the data that we fill inside the cells are relevant.

For example, in TIPS dataframe, TOTAL_BILL is sort of label / variable and its values in the cell aren't an actual variable in order to get this into matrix form.

To do that:

```python
tc = tips.corr()
```

```python
tc
```

Now we have a matrix form of some sort of data here, i.e, the column & the index variable names actually indicate a relevancy to both the column & the index name for the cell where they match up.

```python
sns.heatmap(tc)
```

It colours in those values based on some gradient scale.

```python
sns.heatmap(tc, annot = True)
```

<b>● annot</b> is used to annotate the actual value that belongs to these cells

```python
sns.heatmap(tc, annot = True, cmap='coolwarm')
```

<b>● cmap</b> is used for the colour mapping you want like coolwarm, plasma, magma etc.

```python
sns.heatmap(tc, annot = True, cmap='magma', linewidth=5)
```

<b>● linewidth</b> is used to set the width of the lines separating the cells.

```python
sns.heatmap(tc, annot = True, cmap='magma', linewidth=5, linecolor='black')
```

<b>● linecolor</b> is used to set the colour of the lines separating the cells.

<b>Conclusion:<br></b>
So we can say that all a heatmap does is <b>color the cells based on the gradient and uses some parameters to increase the data visualization. 

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
fp = flights.pivot_table(index='month',columns='year', values='passengers')
```

```python
sns.heatmap(fp)
```

```python
sns.heatmap(fp, cmap='magma')
```

## 2. Cluster Maps

Cluster maps use <b>hierarchical clustering.</b><br> It performs the clustering based on the <b>similarity of the rows and columns.</b>

```python
sns.clustermap(fp) 
```

```python
sns.clustermap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap = 'coolwarm', standard_scale=1) 
```

<b>Clustermap after using standard scaling 
standard_scale = 1</b> normalizes the data from 0 to 1 range.<br>We can see that the months as well as years are no longer in order as they are clustered according to the similarity in case of clustermaps.<br>
So we can conclude that:<br><b>a heatmap will display things in the order we give</b> whereas <br><b>the cluster map clusters the data based on similarity.</b> 

# D) Grids

Grids are general types of plots that allow us to map plot types to grid rows and columns, which helps us to create similar character-separated plots.

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

## 1. Pair Plot

Pair Plot is like an <b>automated joint plot for the entire dataset.</b><br> In a sample, a pair plot maps a <b>pairwise partnership.</b><br>The <b>pairplot() method</b> generates an Axes map, such that each data vector is spread over a single row in the y-axis and across a single column in the x-axis

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue="species", palette="rainbow")
```

<b>● hue</b> - to visualize independent color for each type of ‘species’ in the plot.<br>
<b>● palette</b> - for customizing colors of the plot as shown above

## 2. Pair Grid

We can customize pair plot by using seaborn’s PairGrid mechanism. <br>PairGrid takes all the numerical columns and grids them up making subplots as shown below.

```python
sns.PairGrid(data = iris)
```

We can map to the grid by calling ‘.map’ off of it. Over here we have called the scatter plot by ‘.scatter’. Now all the grids will be of scatter plot kind. We can define the type of plot we want in the grids using the ‘.map’.

```python
g = sns.PairGrid(iris)
g.map(plt.scatter) #All the grids plot scatter plot
```

To map into the upper grid, lower grid and diagonal grid we call <b>.map_upper, .map_lower, .map_diag</b> off of the PairGrid.

```python
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

Hence, the <b>diagonals are hist type plot, the upper grid is’scatter’ type plot and the lower grid is ‘kde’ type plot.

## 3. Facet Grid

FacetGrid is a general way of creating plot grids based on a function.<br>Its object uses the dataframe as input and the names of the variables that shape the row, column, or color dimensions of the grid. 

```python
g = sns.FacetGrid(tips, col="time", row="smoker")
```

The total bills are plotted as hist across the grid we created using <b>map.

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

Over here we have defined <b>hue as sex </b>and also plotted a <b>scatter plot</b> where X-axis is total_bill and Y-axis is yip..

```python
g = sns.FacetGrid(tips, col="time",  row="smoker", hue='sex')
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```

## 4. Joint Grid

JointGrid is the general version for the grid type of jointplot().<br>Jointplot by Seaborn shows a <b>relationship in the margins between 2 variables (bivariate) and 1D profiles (univariate)</b>.<br>This plot is a <b>product form that wraps up the JointGrid.

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.histplot)
```
