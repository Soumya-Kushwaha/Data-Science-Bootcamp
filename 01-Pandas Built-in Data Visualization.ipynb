## The Data

There are some fake data csv files we can read in as dataframes:

## Style Sheets

Matplotlib has style sheets we can use to make our plots look a little nicer. These style sheets include <b>plot_bmh, plot_fivethirtyeight, plot_ggplot and more.</b> <br>They basically create a set of style rules that our plots follow. They make all our plots have the same look and feel more professional. 

Here is how to use them.

**Before plt.style.use() our plots look like this:**

```python
df1['A'].hist()
```

**Call the style:**

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

Now our plots look like this:

```python
df1['A'].hist()
```

```python
plt.style.use('bmh')
df1['A'].hist()
```

```python
plt.style.use('dark_background')
df1['A'].hist()
```

```python
plt.style.use('fivethirtyeight')
df1['A'].hist()
```

```python
plt.style.use('ggplot')
```

Let's stick with the ggplot style and actually see how to utilize pandas built-in plotting capabilities!

# Plot Types

There are several plot types built-in to pandas, most of them statistical plots by nature:

* df.plot.area     
* df.plot.barh     
* df.plot.density  
* df.plot.hist     
* df.plot.line     
* df.plot.scatter
* df.plot.bar      
* df.plot.box      
* df.plot.hexbin   
* df.plot.kde      
* df.plot.pie

We can also just call df.plot(kind='hist') or replace that kind argument with any of the key terms shown in the list above (e.g. 'box','barh', etc..)
___

Let's start going through them!

## Area

```python
df2.plot.area(alpha=0.4)
```

## Barplots

```python
df2.head()
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

## Histograms

```python
df1['A'].plot.hist(bins=50)
```

## Line Plots

```python
df1.plot.line(y='B', figsize=(12,3), lw=1)
```

## Scatter Plots

```python
df1.plot.scatter(x='A',y='B')
```

We can use c to color based off another column value
Use cmap to indicate colormap to use. 

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```

Or use s to indicate size based off another column. s parameter needs to be an array, not just the name of a column:

```python
df1.plot.scatter(x='A',y='B')
```

## BoxPlots

```python
df2.plot.box() # Can also pass a by= argument for groupby
```

## Hexagonal Bin Plot

Useful for Bivariate Data, alternative to scatterplot:

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
df.plot.hexbin(x='a',y='b',gridsize=25,cmap='Oranges')
```
